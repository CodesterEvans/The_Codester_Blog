## The largest issue with following the tutorial

When you go to import the fast.ai library into your code with "from fast.ai.vision.all import *" you will often receive a callback error or some sort of error with the word "callback" in it.

To solve this I simply went to the top of the Collab notebook, clicked "Kernel", and then "Restart Runtime". You may need to do this after not accessing the Collab notebook for a while. It is a strange error but it has always gone away when I restart the runtime. As I understand it, this serves the function of restarting the GPU/notebook.

Voila! You can now rerrun the lines of code to import the fast.ai vision library and have fastai2 ready to go on your computer! This can be confusing because fastai2 goes under the shortened name which is just "fastai" so beware. I ran the cell with "DataBlocks" as shown in the tutorial and it ran without a hitch!

## Ignore The Tutorial's Custom Function For Data_Loaders 

In the tutorial, David Yang defines a valid_func and a label_func which he uses to customize his DataBlock. I tried to understand what he did but it threw me off. Maybe you are a better man than me; however, I simply followed the way Jeremy Howard formatted the DataBlock in the Deep Learning course. Below is the code I used:

1) Ignore David Yang's entire 4th cell with the custom functions.
2) 
db = DataBlock(blocks =(ImageBlock,CategoryBlock),
                 get_items=get_image_files,
                 splitter=RandomSplitter(valid_pct=0.2, seed=42),
                 get_y = parent_label,
                 item_tfms=Resize(sz*2),
                 batch_tfms = [*aug_transforms(size=sz), Normalize.from_stats(*stats)]
)
data = db.dataloaders(data_dir)
data.show_batch(max_n = 8)

This will load 8 chest xray images as well if done successfully.

**note: make sure in the dataloaders function you use "data_dir" instead of "data_path" .
